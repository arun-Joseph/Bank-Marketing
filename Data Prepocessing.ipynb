# Data Prepocessing

```python
import numpy as np
import pandas as pd
import csv
from sklearn import preprocessing
```

```python
with open('bank/bank-full.csv', 'r') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=';')
    df = list(spamreader)
    df = np.asarray(df)
    df = pd.DataFrame(data=df[1:], columns=df[0])
```

```python
numeric = ['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']
df[numeric] = df[numeric].astype(int)

categorical = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome', 'y']
for c in categorical:
    df[c] = pd.factorize(df[c])[0]
    
df = df.astype(float)
```

```python
scaler = preprocessing.MinMaxScaler(feature_range=(0, 1))
#scaler = preprocessing.StandardScaler()
rescaled_df = scaler.fit_transform(df)
df = pd.DataFrame(rescaled_df, columns=df.columns)
```
